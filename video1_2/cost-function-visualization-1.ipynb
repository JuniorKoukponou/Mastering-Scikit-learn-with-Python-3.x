```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def hypothesis(theta1, x):
    return theta1 * x

theta1 = [.1, .4, .8, 1., 1.4, 1.8]

x = np.arange(0, 10, 1)
y = np.arange(0, 10, 1)
Y = [[hypothesis(theta1[i], x_i) for x_i in x] for i in range(len(theta1))]
```

```python
plt.figure()
plt.figure(figsize=(20, 10))
plt_loc = [231, 232, 233, 234, 235, 236]
for i in range(len(theta1)):
    plt.subplot(plt_loc[i])
    plt.scatter(x, y, marker='+', c='red')
    plt.plot(x, Y[i])
    plt.xlim((0, 10))
    plt.ylim((0, 10))
    plt.title('theta1 = ' + str(theta1[i]))
    plt.xlabel('x')
    plt.ylabel('y')
plt.show()
```

```python
J_theta1 = [np.sum(np.square(y - Y[i])) / (2 * len(y)) for i in range(len(theta1))]

plt.figure()
plt.plot(theta1, J_theta1, 'r*-')
plt.title('Cost function')
plt.xlabel('theta1')
plt.ylabel('J_theta1')
plt.show()
```
